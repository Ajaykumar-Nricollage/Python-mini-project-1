```python
questions = [
    {
        "question": "who scored highest centuries in odi's?",
        "options": ["A.Sachin", "B. Gill", "C. Kohli", "D. Rohit"],
        "answer": "C"
    },
    {
        "question": "In which year India got the independence?",
        "options": ["A. 1947", "B. 1963", "C. 1995", "D. 1940"],
        "answer": "A"
    },
    {
        "question": "what is the full form of www?",
        "options": ["A.world wide work", "B. World wide will", "C. World wide web", "D. World wide which"],
        "answer": "C"
    },
    {
        "question": "who is the present prime minister of india?",
        "options": ["A. Rahul gandhi", "B.kcr", "C. jagan", "D.narendra modi "],
        "answer": "D"
    },
    {
        "question": " In which year Ipl was started?",
        "options": ["A. 2000", "B. 2007", "C. 2005", "D. 2008"],
        "answer": "D"
    },
    {
        "question": "who invented 0 in mathematics?",
        "options": ["A.thomas alva edison", "B. Tenali rama krishna", "C. Arya bhatta", "D. Robert"],
        "answer": "C"
    },
    {
        "question": "Who is the present finance minister of india?",
        "options": ["A.Nirmala sitharaman", "B. Arun jaitely", "C. Nimmala ramanayudu", "D.Nara Chandhra baba nayudu"],
        "answer": "A"
    },
    {
        "question": "what is the capital of india?",
        "options": ["A.andhrapradesh", "B. Tamilnadu", "C. Kolkata", "D.Delhi"],
        "answer": "D"
    },
    {
        "question": "which country odi world Cup 2023 was held?",
        "options": ["A. America", "B. India", "C. Australia", "D. England"],
        "answer": "B"
    },
    {
        "question": "what is the colour of sky?",
        "options": ["A. Red", "B. White", "C. Green", "D. None of these"],
        "answer": "D"
    }
]

def quiz_game():
    print("Welcome to the Quiz Game!")
    print("Answer the questions by typing the letter of the correct option (A, B, C, or D).\n")

    score = 0
    question_number = 1

    for q in questions:
        print(f"Question {question_number}: {q['question']}")

        for option in q["options"]:
            print(option)

        answer = input("Your answer: ")

        if answer == q["answer"]:
            print("Correct!\n")
            score += 10
        else:
            print(f"Wrong! The correct answer was {q['answer']}.\n")

        question_number += 1

    print(f"Quiz Over! Your total score is: {score}/100")

    if score == 100:
        print("Excellent! Your are awesome!")
    elif score >= 50:
        print("Good!")
    else:
        print("Better luck next time!")

if __name__ == "__main__":
    quiz_game()
```
